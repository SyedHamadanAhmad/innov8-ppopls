```python
from sklearn.feature_extraction.text import CountVectorizer

# Function to add missing columns and populate them
def add_missing_columns(df, df_test):
    # Identify columns that are in df but not in df_test, excluding 'species'
    cols_in_df_not_in_test = set(df.columns) - set(df_test.columns) - {'species'}
    cols_in_test_not_in_df = set(df_test.columns) - set(df.columns)

    # Add missing columns to df_test
    for col in cols_in_df_not_in_test:
        df_test[col] = 0  # Initialize with 0

    # Add missing columns to df
    for col in cols_in_test_not_in_df:
        if col != 'species':  # Skip 'species' when adding to df
            df[col] = 0  # Initialize with 0

    # Transform 'message' column to create word presence for df
    vectorizer = CountVectorizer(binary=True)
    word_matrix_df = vectorizer.fit_transform(df['message'])
    word_df = pd.DataFrame(word_matrix_df.toarray(), columns=vectorizer.get_feature_names_out())

    # Set the appropriate columns in df based on the presence in 'message'
    for col in word_df.columns:
        if col in df.columns:
            df[col] = word_df[col]

    # Transform 'message' column to create word presence for df_test
    word_matrix_test = vectorizer.transform(df_test['message'])
    word_df_test = pd.DataFrame(word_matrix_test.toarray(), columns=vectorizer.get_feature_names_out())

    # Set the appropriate columns in df_test based on the presence in 'message'
    for col in word_df_test.columns:
        if col in df_test.columns:
            df_test[col] = word_df_test[col]

    return df, df_test

# Apply the function
df, df_test = add_missing_columns(df, df_test)

# Show the updated DataFrames
print("Updated df columns:", df.columns)
print("Updated df_test columns:", df_test.columns)
```

```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier

# Separating features (X) by dropping the 'species' column
X = df.drop(columns=['species'])

# Target column (y) remains the 'species' column
y = df['species']

# Initialize the K-Nearest Neighbors classifier
knn = KNeighborsClassifier(n_neighbors=3)  # k=1 as specified

# Train the KNN model on the entire dataset
knn.fit(X, y)

# Prepare the test features from df_test
X_test = df_test  # Ensure df_test has the same feature columns

# Make predictions on df_test
y_test_pred = knn.predict(X_test)

# Create a DataFrame to store predictions
df_test['predicted_species'] = y_test_pred

# Display the predictions
print(df_test[['predicted_species']])  # Assuming df_test has a 'species' column for reference
```
