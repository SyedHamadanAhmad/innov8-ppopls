```python
from sklearn.feature_extraction.text import CountVectorizer

# Function to add missing columns and populate them
def add_missing_columns(df, df_test):
    # Identify columns that are in df but not in df_test, excluding 'species'
    cols_in_df_not_in_test = set(df.columns) - set(df_test.columns) - {'species'}
    cols_in_test_not_in_df = set(df_test.columns) - set(df.columns)

    # Add missing columns to df_test
    for col in cols_in_df_not_in_test:
        df_test[col] = 0  # Initialize with 0

    # Add missing columns to df
    for col in cols_in_test_not_in_df:
        if col != 'species':  # Skip 'species' when adding to df
            df[col] = 0  # Initialize with 0

    # Transform 'message' column to create word presence for df
    vectorizer = CountVectorizer(binary=True)
    word_matrix_df = vectorizer.fit_transform(df['message'])
    word_df = pd.DataFrame(word_matrix_df.toarray(), columns=vectorizer.get_feature_names_out())

    # Set the appropriate columns in df based on the presence in 'message'
    for col in word_df.columns:
        if col in df.columns:
            df[col] = word_df[col]

    # Transform 'message' column to create word presence for df_test
    word_matrix_test = vectorizer.transform(df_test['message'])
    word_df_test = pd.DataFrame(word_matrix_test.toarray(), columns=vectorizer.get_feature_names_out())

    # Set the appropriate columns in df_test based on the presence in 'message'
    for col in word_df_test.columns:
        if col in df_test.columns:
            df_test[col] = word_df_test[col]

    return df, df_test

# Apply the function
df, df_test = add_missing_columns(df, df_test)

# Show the updated DataFrames
print("Updated df columns:", df.columns)
print("Updated df_test columns:", df_test.columns)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix



# Separating features (X) by dropping the 'species' column
X = df.drop(columns=['species'])

# Target column (y) remains the 'species' column
y = df['species']

# Label encode the target variable 'species'
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split the dataset into training and test sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Initialize the K-Nearest Neighbors classifier
knn = KNeighborsClassifier(n_neighbors=1)  # k=3 as an example

# Train the KNN model
knn.fit(X_train, y_train)

# Make predictions on the test data
y_test_pred = knn.predict(X_test)

# Generate and print classification report
print("Classification Report:")
print(classification_report(y_test, y_test_pred, target_names=label_encoder.classes_))

# Generate and print confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_test_pred))

# Analyze the results to identify and handle any low-performing classes
print("\nAnalysis of Low-Performing Classes:")
class_report = classification_report(y_test, y_test_pred, output_dict=True)

# Create a list of low-performing classes based on F1-score
low_performing_classes = {label: metrics for label, metrics in class_report.items()
                          if label != 'accuracy' and metrics['f1-score'] < 0.7}

print("Classes with F1-score < 0.6:")
for cls, metrics in low_performing_classes.items():
    print(f"{cls}: Precision = {metrics['precision']:.2f}, Recall = {metrics['recall']:.2f}, F1-score = {metrics['f1-score']:.2f}")
```
